# Leitura e plotagem de imagens utilizando Python

```python
import matplotlib.pyplot as plt   # Biblioteca utilizada para plotar figuras
                                  # Pode ser usada tamb√©m para ler figuras do disco
import cv2 as cv                  # Biblioteca OpenCV, muito utilizada em processamento de imagens
```

```python
# Leitura da imagem utilizando o matplotlib
img = plt.imread('earth.jpg')

# Leitura da imagem utilizando o OpenCV
img_cv = cv.imread('earth.jpg')
```

```python
# Plotagem da imagem
plt.imshow(img)
```

```python
# Mostra cada canal de cor da imagem separadamente
img_red = img[:,:,0]
img_green = img[:,:,1]
img_blue = img[:,:,2]

plt.figure(figsize=[15, 5.])
plt.subplot(1, 3, 1)
plt.imshow(img_red, cmap='gray')
plt.title('Red channel')
plt.subplot(1, 3, 2)
plt.imshow(img_green, cmap='gray')
plt.title('Green channel')
plt.subplot(1, 3, 3)
plt.imshow(img_blue, cmap='gray')
plt.title('Blue channel')
plt.tight_layout()
```

```python
# Converte imagem colorida para n√≠vel de cinza
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(img_gray, 'gray')
```
